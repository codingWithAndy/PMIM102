### __1. Using a Remote 'git' Repository: <i>github</i>__

'github' is a website which hosts a huge number of online git repositories. It is a very useful way to collaborate with other data scientists on code. We can all take a copy of the code and work on it at our leisure on our own PC (maintaining local version control with git). When we are ready to share our modified code with our collaborators we can copy our changes back to github using a process called __pushing__ the code.

If we want to update the local copy on oour PC that we are working on with changes that someone else has pushed to github, we do this by __pulling__ the updated code from github.

You are likely to find that when you come to __push__ someone else has already made a change and you will need to __pull__ and maybe __merge__ these changes into your code before github will let you __push__.

To start with you are going to make a copy of the files I have gathered in an online git repository.

#### _2.1 Matrices and Contour Plots Example_

There is a nice output from this sample of code and it's not too complex. Matrices don't crop up all that often but it is good to know that they are there.

Experiment with the code to get used to matrices and mathematical functions as well as a few details about producing R code.


```r
# ------------------------------------------------------------------------------
# Matrices and contours
# Persian Rug Art
# ------------------------------------------------------------------------------
# outer is also possible as %o% compared to inner product %*%.
# also have kronecker
# ------------------------------------------------------------------------------
# Create vectors to provide the values for x and y calculations.
x <- y <- seq(-4*pi, 4*pi, len = 27)
# Calculate a function using the outer product of x and y and the function '+'.
r <- sqrt(outer(x^2, y^2, "+"))
# Visualise the values as a filled contour plot.
filled.contour(r, #axes = FALSE,
               color.palette = terrain.colors,
               frame.plot = FALSE, plot.axes = {})

# Run through a series of ordinary contour plots for different transformations.
par(mfrow = c(2, 2), mar = rep(0, 4))
for(f in pi^(0:3))
    contour(g <- cos(r^2)*exp(-r/f),
            drawlabels = FALSE, axes = FALSE, frame = TRUE)

# Try a random variable and solve it to see what we get.
par(mfrow=c(1, 1), mar=rep(1, 4))
s <- matrix(rbinom(27 * 27, size=10, prob=0.5), nrow=27, ncol=27)
t <- solve(s)
filled.contour(t, #axes = FALSE,
               color.palette = terrain.colors,
               frame.plot = FALSE, plot.axes = {})
```

We are told in the documentation that the data are the monthly lung disease deaths since January 1974, so we can add a date for each entry. But then we need to concern ourselves with what a month is?

We will have a look at the Tidyverse library, lubridate, later which makes this easier but you still need to be aware of the problem.

```r
# Add the dates.
df <- cbind(Date=as.Date(0, origin='1974-01-01'), df)
for(i in 2:nrow(df)){
    df[i, 1] <- df[i-1, 1] + 30
}
head(df)
df <- cbind(Date=as.POSIXlt('1974-01-01'), df)
for(i in 2:nrow(df)){
    #browser()
    d <- as.POSIXlt(df[i-1, 1])
    #print(class(d))
    d$mon <- d$mon + 1
    # print(d)
    df[i, 1] <- as.POSIXct(d)
}
head(df)
df <- cbind(Date=seq(from = as.Date("1974-01-01"), to = as.Date("1979-12-31"), by = 'month'), df)
head(df)

plot(df[, c(1,4:6)])
```

```r
##install.packages("faraway")
#library(faraway)
#?faraway::prostate
```

```r
# ------------------------------------------------------------------------------
# Prostate Cancer
# ------------------------------------------------------------------------------
# Format
# This data frame contains the following columns:
# lcavol - log(cancer volume) 
# lweight - log(prostate weight) 
# age - age 
# lbph - log(benign prostatic hyperplasia amount) 
# svi - seminal vesicle invasion 
# lcp - log(capsular penetration) 
# gleason - Gleason score 
# pgg45 - percentage Gleason scores 4 or 5 
# lpsa - log(prostate specific antigen)
#
# install.packages('faraway')

library(faraway)
attach(prostate)

print(head(prostate))

summary(prostate)

options(repr.plot.width = 8, repr.plot.height = 8)
plot(prostate)

# Plot of PSA levels with gleason.
options(repr.plot.width = 4, repr.plot.height = 4)
boxplot(lpsa ~ gleason)

# Tables svi vs gleason
table(svi, gleason)

# Create a factor for age-group.
par(mfrow=c(1,2))
options(repr.plot.width = 8, repr.plot.height = 4)
prostate$agegp <- as.factor(ifelse(age<50, '<50', ifelse(age<60, '50-60', ifelse(age<70, '60-70', '>70'))))
with(prostate, boxplot(lpsa ~ agegp))
prostate$agegp <- factor(prostate$agegp, levels=c('<50', '50-60', '60-70', '>70'))
with(prostate, boxplot(lpsa ~ agegp))

# Get psa range (< 10, 10-20, >20).
par(mfrow=c(1,2))
prostate$psagp <- as.factor(ifelse(lpsa<log(10), '<10', ifelse(lpsa<log(20), '10-20', '>20')))
with(prostate, boxplot(lpsa ~ psagp))
prostate$psagp <- factor(prostate$psagp, levels=c('<10', '10-20', '>20'))
with(prostate, boxplot(lpsa ~ psagp))

par(mfrow=c(1,2))
plot(exp(lpsa), exp(lcavol))
plot(exp(lpsa), lcavol)

par(mfrow=c(1,2))
qqnorm(exp(lpsa))
qqline(exp(lpsa))

qqnorm(lpsa)
qqline(lpsa)

shapiro.test(lpsa)

# Simple linear regression.
par(mfrow=c(1,2))
m<-lm(lcavol~lpsa, data=prostate)
plot(lpsa,lcavol)
abline(m, col='red')

m<-lm(lcavol~exp(lpsa), data=prostate)
plot(exp(lpsa),lcavol)
abline(m, col='green')
```
